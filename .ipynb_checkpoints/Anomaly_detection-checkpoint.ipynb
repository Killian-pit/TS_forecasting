# Time series anilomaly detection using multiple techniques 

* Dense autoencoder  
* LSTM Autoencoder system  


```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler

from  tensorflow.keras import Sequential
from  tensorflow.keras.layers import Dropout, Dense,TimeDistributed,LSTM,RepeatVector
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
```

```python
def create_dataset(X, time_steps=1):
    Xs= []
    for i in range(len(X) - time_steps):
        v = X[i:(i + time_steps)]
        Xs.append(v)
    return np.array(Xs)

# definition of the network usiing the sequential API 

def autoencoder_lstm(x):
    # define model
    model = Sequential()
    model.add(LSTM(8, activation='linear', input_shape=(x.shape[1],x.shape[2])))
    model.add(RepeatVector(x.shape[1]))
    model.add(LSTM(8, activation='linear', return_sequences=True))
    model.add(TimeDistributed(Dense(x.shape[2])))
    model.compile(optimizer='adam', loss='mae')
    return model 

# create the dense autoencoder model using the functional api 
def autoencoder_dense(x):
    # define model
    
    input_ts = Input(shape=(x.shape[1],))
    encoder = Dense(8,activation = 'linear')(input_ts)
    decoder = Dense(x.shape[1],activation='linear')(encoder)
    model  = Model(input_ts,decoder)
    model.compile(optimizer='adam', loss='mse')
    return model 



```

```python
# variables 

TIME_STEPS = 5
```

```python
# create an artifical dataset

x = np.linspace(0,365,5*365)
y = np.cos(x)
y = y + .001*y*x
y = y + .01*x
y = y + np.random.uniform(-.1,.1,len(y))

y[-100:-40] = y[-100:-40] + y[-100:-40]*np.random.uniform(0,.2,len(y[-100:-40]))

plt.figure(figsize=(20,10))
plt.plot(x,y)
```

```python
train_size = int(len(y) * 0.80)
test_size = len(y) - train_size
train, test = y[0:train_size], y[train_size:len(y)]
print(train.shape, test.shape)
```

# Dense autoencoder

```python
train_size = int(len(y) * 0.80)
test_size = len(y) - train_size
train, test = y[0:train_size], y[train_size:len(y)]
print(train.shape, test.shape)

scaler = StandardScaler()
scaler = scaler.fit(train.reshape(-1, 1))
train = scaler.transform(train.reshape(-1, 1))
test  = scaler.transform(test.reshape(-1, 1))
```

```python
TIME_STEPS = 5

test_size = len(y) - train_size
train, test = y[0:train_size], y[train_size:len(y)]
print(train.shape, test.shape)

scaler = StandardScaler()
scaler = scaler.fit(train.reshape(-1, 1))


X_train = create_dataset(train,
                         TIME_STEPS
                        )

X_test = create_dataset(test,
                        TIME_STEPS
                       )

print(X_train.shape)
print(X_test.shape)
```

```python
model = autoencoder_dense(X_train)
model.summary()


history = model.fit(
    X_train, X_train,
    epochs=50,
    batch_size=32,
    validation_split=0.1,
    shuffle=False
)
```

```python
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

```python
X_train_pred = model.predict(X_train)
X_test_pred = model.predict(X_test)

train_mae_loss = np.mean(np.abs(X_train_pred - X_train), axis=1)
test_mae_loss = np.mean(np.abs(X_test_pred - X_test), axis=1)

plt.figure()
_ = plt.hist(train_mae_loss,bins=100)
plt.title('Training error distribution')
plt.figure()
plt.title('Test error distribution')
_ = plt.hist(test_mae_loss,bins=100)
```

```python
THRESHOLD = 0.2

plt.plot(x[-len(test[TIME_STEPS:]):],test[TIME_STEPS:])
a = np.where(test_mae_loss > THRESHOLD,test[TIME_STEPS:],np.nan)
plt.scatter(x[-len(test[TIME_STEPS:]):],
            a,c= 'r')
plt.title('Anomaly detection using dense autoencoder')
```

# LSTM autoencoder 

```python
# preprocess the input data to be feed to the autoencoder 
# need to be 3 d 

train_size = int(len(y) * 0.80)
test_size = len(y) - train_size
train, test = y[0:train_size], y[train_size:len(y)]
print(train.shape, test.shape)

scaler = StandardScaler()
scaler = scaler.fit(train.reshape(-1, 1))
train = scaler.transform(train.reshape(-1, 1))
test  = scaler.transform(test.reshape(-1, 1))
```

```python
# reshape to [samples, time_steps, n_features]

X_train = create_dataset(
  train,
  TIME_STEPS
)

X_test = create_dataset(
  test,
  TIME_STEPS
)

print(X_train.shape)
print(X_test.shape)
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1],1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1],1)
```

```python
model = autoencoder_lstm(X_train)
model.summary()

history = model.fit(
    X_train, X_train,
    epochs=100,
    batch_size=8,
    validation_split=0.1,
    shuffle=False
)
```

```python
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

```python
X_train_pred = model.predict(X_train)
X_test_pred = model.predict(X_test)

train_mae_loss = np.mean(np.abs(X_train_pred - X_train), axis=1)
test_mae_loss = np.mean(np.abs(X_test_pred - X_test), axis=1)

plt.figure()
_ = plt.hist(train_mae_loss,bins=100)
plt.title('Training error distribution')
plt.figure()
plt.title('Test error distribution')
_ = plt.hist(test_mae_loss,bins=100)
```

```python
THRESHOLD = 0.15
plt.plot(x[-len(test[TIME_STEPS:]):],test[TIME_STEPS:])
plt.scatter(x[-len(test[TIME_STEPS:]):],
            np.where(test_mae_loss > THRESHOLD,test[TIME_STEPS:],np.nan),
            c= 'r')
```

# Using tensorflow probability 

```python
import tensorflow_probability as tfp 
import tensorflow as tf 
import tensorflow.keras as tfk
tfd = tfp.distributions
```

```python
encoded_size = 4
prior = tfd.Independent(tfd.Normal(loc=tf.zeros(encoded_size), scale=1),
                        reinterpreted_batch_ndims=1)
```

```python
# preprocess the input data to be feed to the autoencoder 
# need to be 3 d 

train_size = int(len(y) * 0.80)
test_size = len(y) - train_size
train, test = y[0:train_size], y[train_size:len(y)]
print(train.shape, test.shape)

scaler = StandardScaler()
scaler = scaler.fit(train.reshape(-1, 1))


X_train = create_dataset(train,
                         TIME_STEPS
                        )

X_test = create_dataset(test,
                        TIME_STEPS
                       )

print(X_train.shape)
print(X_test.shape)
```

```python
# create the network 

tfpl = tfp.layers
tfkl = tfk.layers
input_shape = X_train.shape[1]

encoder = Sequential([
    tfkl.InputLayer(input_shape=input_shape),
    #tfkl.Lambda(lambda x: tf.cast(x, tf.float32) - 0.5),
    tfkl.Dense(32,activation=tf.nn.leaky_relu),
    tfkl.Dense(16,activation=tf.nn.leaky_relu),
    tfkl.Dense(tfpl.MultivariateNormalTriL.params_size(encoded_size),
               activation=None),
    tfpl.MultivariateNormalTriL(
        encoded_size,
        activity_regularizer=tfpl.KLDivergenceRegularizer(prior, weight=1.0)),
])

decoder = Sequential([
    tfkl.InputLayer(input_shape=[encoded_size]),
    tfkl.Dense(16,activation=tf.nn.leaky_relu),
    tfkl.Dense(8,activation=tf.nn.leaky_relu),
    tfpl.IndependentBernoulli(input_shape, tfd.Bernoulli.logits),
])

vae = Model(inputs=encoder.inputs,
                outputs=decoder(encoder.outputs))

negative_log_likelihood = lambda x, rv_x: -rv_x.log_prob(x)

vae.compile(optimizer=tf.optimizers.Adam(learning_rate=1e-3),
            loss=negative_log_likelihood)

vae.fit(X_train, X_train,
    epochs=100,
    batch_size=2048,
    validation_split=0.1,
    shuffle=False)
```
